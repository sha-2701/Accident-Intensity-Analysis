```python
d = {}
for i in dataset['Start_Time']:
    if i[:4] in d:
        d[i[:4]] += 1
    else:
        d[i[:4]] = 1 
sd = list(d.keys())
sd.sort()
d_sort1 = {i: d[i] for i in sd}
```

```python
d1 = {}
for i in dataset['End_Time']:
    if i[:4] in d1:
        d1[i[:4]]+= 1
    else:
        d1[i[:4]] = 1
sd = list(d.keys())
sd.sort()
d_sort2 = {i: d[i] for i in sd}
dic= [d_sort1,d_sort2]
```

```python
j = 0
for i in datetime1:
    plt.figure(figsize=(28, 6))
    plt.grid()
    x = dic[j].keys()
    y = dic[j].values()
    j += 1
    plt.plot(x, y)
    plt.xlabel(i)
    plt.ylabel("Number of accidents")
    plt.show()
```

<h1>Geopandas</h1>

```python
import geopandas as gpd
%matplotlib inline
from shapely.geometry import Point
```

```python
dataset.head()
```

```python
point = dataset.apply(lambda row: Point(row.Start_Lng, row.Start_Lat), axis=1)
```

```python
geodataset = gpd.GeoDataFrame(dataset, geometry=point)
geodataset.crs = {'init':'epsg:4326'}
```

```python
us = gpd.read_file("cb_2018_us_state_5m.shp")
usa =us.plot(figsize=(20,20),edgecolor='black',legend = True ,linewidth = -0.5)
usa.set_xlim([-130,-60])
usa.set_ylim([20,50])
```

```python
us.to_crs(geodataset.crs, inplace= True)
```

```python
geodataset.crs
```

<hr>
In Spatial join, We can do join as how='inner' and how='left' <br>
op ='within' when we join dataset to states <br>
op = 'contains' when we join states to dataset
<hr> 

```python
dataset_with_join = gpd.sjoin(us,geodataset,how = 'left', op ='contains')
```
